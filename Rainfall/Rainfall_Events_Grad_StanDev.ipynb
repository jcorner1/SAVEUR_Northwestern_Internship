```python
#Import Mods
import pandas as pd
import numpy as np

from datetime import datetime
from datetime import timedelta

import glob

from math import radians, cos, sin, asin, sqrt
from scipy import stats
```

```python
#Defintion from geeksforgeeks.org
#Python 3 program to calculate Distance Between Two Points on Earth from lat-lon to kilometer.
def distance(lat1, lat2, lon1, lon2): 
      
    # The math module contains a function named 
    # radians which converts from degrees to radians. 
    lon1 = radians(lon1) 
    lon2 = radians(lon2) 
    lat1 = radians(lat1) 
    lat2 = radians(lat2) 
       
    # Haversine formula  
    dlon = lon2 - lon1  
    dlat = lat2 - lat1 
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
  
    c = 2 * asin(sqrt(a))  
     
    # Radius of earth in kilometers. Use 3956 for miles 
    r = 6371
       
    # calculate the result 
    return(c * r) 
      
      
```

```python
#Array for all the data
array_2d = []

#Open files within full folder
for filename in glob.glob("C:/Users/jerem/OneDrive/Desktop/SAVEUR/SAVEUR Internship/Data/Rainfall_Events/*.csv"):

    
    file = pd.read_csv(filename)
    
    #Save off columns for files to be used later
    stations = file.station
    rainfall = file.rain_total
    start_date = file.start_date
    end_date = file.end_date

    #reset rain totals for each station and array
    rain_ORD = 0
    rain_JOT = 0
    rain_PWK = 0
    rain_UGN = 0
    rain_DPA = 0
    rain_ARR = 0
    rain_LOT = 0
    rain_IGQ = 0
    array_1d = []


    for i in range(len(stations)):
        if stations[i] == "ORD":
            rain_ORD = rain_ORD + rainfall[i]
        elif stations[i]== "JOT":
            rain_JOT = rain_JOT + rainfall[i]
        elif stations[i]== "PWK":
            rain_PWK = rain_PWK + rainfall[i]
        elif stations[i]== "UGN":
            rain_UGN = rain_UGN + rainfall[i]
        elif stations[i]== "DPA":
            rain_DPA = rain_DPA + rainfall[i]
        elif stations[i]== "ARR":
            rain_ARR = rain_ARR + rainfall[i]
        elif stations[i]== "LOT":
            rain_LOT = rain_LOT + rainfall[i]
        elif stations[i]== "IGQ":
            rain_IGQ = rain_IGQ + rainfall[i]
            
    Total_Rainfall_List = [rain_ORD, rain_JOT, rain_PWK, rain_UGN, rain_DPA, rain_ARR, rain_LOT, rain_IGQ]
    List = ['ORD', 'JOT', 'PWK', 'UGN', 'DPA', 'ARR', 'LOT', 'IGQ']
    lat = [41.9875, 41.5177, 42.1208, 42.4222, 41.9078, 41.7700, 41.6081, 41.5349]
    lon = [-87.9319, -88.1756, -87.9047, -87.8679, -88.2486, -88.4800, -88.0962, -87.5295]


    if any(i >= 3 for i in Total_Rainfall_List):
        stdev = np.std(Total_Rainfall_List)
        max_value = np.max(Total_Rainfall_List)
        max_loc = np.where(max_value == Total_Rainfall_List)
        location = max_loc[0]
        max_ASOS = List[location[0]]
        max_rain = Total_Rainfall_List[location[0]]
        max_lat = lat[location[0]]
        max_lon = lon[location[0]]
        
        #Create a loop to find the distance between the max and all other ASOS
        #Loop skips value of max location for easier
        dist_array = []
        asos_array = []

        for j in [x for x in range(len(List)) if x != location[0]]:
            lat1 = max_lat
            lat2 = lat[j]
            lon1 = max_lon
            lon2 = lon[j]
            dist_array.append(distance(lat1, lat2, lon1, lon2))
            asos_array.append(List[j])
    
        idx = np.argsort(dist_array)[:3]

        Start = stats.mode(start_date)
        End = stats.mode(end_date)

        array_1d.append((Start.mode[0], End.mode[0], max_rain, max_ASOS, stdev))

        for k in idx:
            if asos_array[k] == 'ORD':
                grad = (max_rain - rain_ORD)/dist_array[k]
                array_1d.append((grad, asos_array[k]))

            elif asos_array[k] == 'JOT':
                grad = (max_rain - rain_JOT)/dist_array[k]
                array_1d.append((grad, asos_array[k]))  

            elif asos_array[k] == 'PWK':
                grad = (max_rain - rain_PWK)/dist_array[k]
                array_1d.append((grad, asos_array[k]))

            elif asos_array[k] == 'UGN':
                grad = (max_rain - rain_UGN)/dist_array[k]
                array_1d.append((grad, asos_array[k]))

            elif asos_array[k] == 'DPA':
                grad = (max_rain - rain_DPA)/dist_array[k]
                array_1d.append((grad, asos_array[k]))

            elif asos_array[k] == 'ARR':
                grad = (max_rain - rain_ARR)/dist_array[k]
                array_1d.append((grad, asos_array[k]))

            elif asos_array[k] == 'LOT':
                grad = (max_rain - rain_LOT)/dist_array[k]
                array_1d.append((grad, asos_array[k]))

            elif asos_array[k] == 'IGQ':
                grad = (max_rain - rain_IGQ)/dist_array[k]
                array_1d.append((grad, asos_array[k]))

        array_2d.append(array_1d)



```
